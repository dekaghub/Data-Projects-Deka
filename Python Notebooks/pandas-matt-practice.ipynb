## Idea of Chain in depth - Flow programming

```python
# easy to read, comment & debug
# use pipe to pass in functions

# Jupyter by default displays the last line, use display to print intermediate or more
# display() over print because print outputs text only whereas display() is html based
from IPython.display import display 

# sample function to assign a df to a global variable
def get_var(df, var_name):
    globals()[var_name] = df
    return df

def clean_data(autos):
    return (
    autos[cols]
        .pipe(get_var, 'df_pipe') # passing in function with argument
        .assign(cylinders=autos.cylinders.fillna(0).astype('int8'),
        displ=autos.displ.fillna(0).astype('float32'),
        drive=autos.drive.fillna('Other').astype('category'),
        automatic=autos.trany.str.contains('Auto'),
        speeds=autos.trany.str.extract(r'(\d)+').fillna('20').astype('int8'),
        createdOn=pd.to_datetime(autos.createdOn.replace({' EDT':'-04:00', ' EST':'-05:00'}, regex=True))
        )
        .pipe(lambda df: display(df) or df) # good for debugging or intermediate df
        .astype({'city08':'int8', 'comb08':'int8', 'highway08':'int8',
                'fuelCost08':'int16', 'range':'int16','year':'int16', 'make':'category'})
        .drop(columns=['trany'])
    )
```
